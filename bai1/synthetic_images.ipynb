```python
import os
import cv2
import numpy as np
from IPython.display import display, Image
os.makedirs('output_images/bai4', exist_ok=True)
```

```python
height, width = 256, 256
gradient = np.zeros((height, width), dtype=np.uint8)
for i in range(width):
    gradient[:, i] = int(i / (width - 1) * 255)
cv2.imwrite('output_images/bai4/gradient.png', gradient)
print('dtype:', gradient.dtype, '| shape:', gradient.shape)
display(Image(filename='output_images/bai4/gradient.png'))
```

```python
checker = np.zeros((height, width), dtype=np.uint8)
cell = 32
for r in range(0, height, cell):
    for c in range(0, width, cell):
        if ((r // cell) + (c // cell)) % 2 == 0:
            checker[r:r+cell, c:c+cell] = 255
cv2.imwrite('output_images/bai4/checkerboard.png', checker)
print('dtype:', checker.dtype, '| shape:', checker.shape)
display(Image(filename='output_images/bai4/checkerboard.png'))
```

```python
circle = np.zeros((height, width), dtype=np.uint8)
cv2.circle(circle, (width // 2, height // 2), 80, 255, -1)
cv2.imwrite('output_images/bai4/circle.png', circle)
print('dtype:', circle.dtype, '| shape:', circle.shape)
display(Image(filename='output_images/bai4/circle.png'))
```
