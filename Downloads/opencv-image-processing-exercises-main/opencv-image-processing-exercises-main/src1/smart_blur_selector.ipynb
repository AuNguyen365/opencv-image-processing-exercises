```python
def gaussian_kernel(k):
    g = cv2.getGaussianKernel(k, -1)
    return g @ g.T
```

```python
def fft_blur(image, kernel):
    h, w = image.shape
    kh, kw = kernel.shape

    pad = np.zeros((h, w))
    pad[:kh, :kw] = kernel

    F = np.fft.fft2(image)
    H = np.fft.fft2(pad)

    G = F * H
    result = np.fft.ifft2(G)

    return np.real(result)
```
