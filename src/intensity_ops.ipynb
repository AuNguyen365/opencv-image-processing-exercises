```python
import cv2
import numpy as np
import matplotlib.pyplot as plt 
```

```python
gray = cv2.imread("input_images/bai1/photo.jpg", cv2.IMREAD_GRAYSCALE)

if gray is None:
    print("Không đọc được ảnh! Hãy kiểm tra lại đường dẫn.")
else:
    print("Đọc ảnh thành công! Kích thước:", gray.shape)
    plt.imshow(gray, cmap='gray') # Hiện ảnh gốc lên xem thử
    plt.show()
```

```python
def clip_uint8(arr):
    return np.clip(arr, 0, 255).astype(np.uint8)
```

```python
# Ảnh tối
gray_dark = clip_uint8(gray.astype(np.int32) - 50)

# Ảnh sáng
gray_bright = clip_uint8(gray.astype(np.int32) + 50)

# Tăng tương phản
gray_contrast = clip_uint8(gray.astype(np.int32) * 1.5)

# Hiển thị thử một cái
plt.imshow(gray_dark, cmap='gray')
plt.title("Anh toi")
plt.show()
```

```python
T = 128
binary = np.where(gray >= T, 255, 0).astype(np.uint8)

plt.imshow(binary, cmap='gray')
plt.title("Anh nhi phan")
plt.show()
```

```python
cv2.imwrite("output_images/bai1/dark.jpg", gray_dark)
cv2.imwrite("output_images/bai1/bright.jpg", gray_bright)


print("Pixel trước:", gray[100,100])
print("Pixel sau khi làm tối:", gray_dark[100,100])
```
