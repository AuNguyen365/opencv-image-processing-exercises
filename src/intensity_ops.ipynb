```python
import cv2
import numpy as np
import matplotlib.pyplot as plt 
```

```python
gray = cv2.imread("input_images/bai2/photo.jpg", cv2.IMREAD_GRAYSCALE)

if gray is None:
    print("Không đọc được ảnh! Hãy kiểm tra lại đường dẫn.")
else:
    print("Đọc ảnh thành công! Kích thước:", gray.shape)
    plt.imshow(gray, cmap='gray')
    plt.show()
```

```python
def clip_uint8(arr):
    return np.clip(arr, 0, 255).astype(np.uint8)
```

```python
gray_dark = clip_uint8(gray.astype(float) - 50)
gray_bright = clip_uint8(gray.astype(float) + 50)
gray_contrast = clip_uint8(gray.astype(float) * 1.5)
```

```python
plt.figure(figsize=(12,4))

# ảnh tối
plt.subplot(1,3,1)
plt.imshow(gray_dark, cmap='gray')
plt.title("Anh toi")
plt.axis('off')

# ảnh sáng
plt.subplot(1,3,2)
plt.imshow(gray_bright, cmap='gray')
plt.title("Anh sang")
plt.axis('off')

# tăng tương phản
plt.subplot(1,3,3)
plt.imshow(gray_contrast, cmap='gray')
plt.title("Tang tuong phan")
plt.axis('off')

plt.show()
```

```python
T = 128
binary = np.where(gray >= T, 255, 0).astype(np.uint8)

plt.imshow(binary, cmap='gray')
plt.title("Anh nhi phan")
plt.show()
```

```python
cv2.imwrite("output_images/bai2/dark.jpg", gray_dark)
cv2.imwrite("output_images/bai2/bright.jpg", gray_bright)
cv2.imwrite("output_images/bai2/contrast.jpg", gray_contrast)
cv2.imwrite("output_images/bai2/binary.jpg", binary)


print("Pixel gốc:", gray[100,100])
print("Pixel tối:", gray_dark[100,100])
print("Pixel sáng:", gray_bright[100,100])
print("Pixel tương phản:", gray_contrast[100,100])
print("Pixel nhị phân:", binary[100,100])
```
