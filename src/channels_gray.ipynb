# Channels & Gray Notebook


```python
import os
import cv2
import numpy as np
from IPython.display import display, Image
os.makedirs('output_images/bai1', exist_ok=True)
```

```python
img = cv2.imread('input_images/bai1/photo.jpg')
if img is None:
    raise FileNotFoundError('Error: Cannot read image photo.jpg')
print('Color image shape:', img.shape)
```

```python
B = img[:, :, 0]
G = img[:, :, 1]
R = img[:, :, 2]
print('Blue channel shape:', B.shape)
print('Green channel shape:', G.shape)
print('Red channel shape:', R.shape)
cv2.imwrite('output_images/bai1/blue.png', B)
cv2.imwrite('output_images/bai1/green.png', G)
cv2.imwrite('output_images/bai1/red.png', R)
```

```python
gray = ((R.astype(float) + G.astype(float) + B.astype(float)) / 3).astype(np.uint8)
print('Gray image shape:', gray.shape)
cv2.imwrite('output_images/bai1/gray_manual.png', gray)
```

```python
display(Image(filename='output_images/bai1/blue.png'))
display(Image(filename='output_images/bai1/green.png'))
display(Image(filename='output_images/bai1/red.png'))
display(Image(filename='output_images/bai1/gray_manual.png'))
```
