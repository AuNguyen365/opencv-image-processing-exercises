```python
import cv2
import os
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
img_path = 'input_images\input.jpg'
img = cv2.imread(img_path)

if img is None:
    print(f"Lỗi: Không tìm thấy ảnh '{img_path}'. Hãy kiểm tra lại thư mục hiện tại.")
else:
    img_half = img[::2, ::2]
    img_quarter = img[::4, ::4]

    print("--- THÔNG TIN KÍCH THƯỚC (SHAPE) ---")
    print(f"Shape ảnh gốc : {img.shape}")
    print(f"Shape ảnh 1/2 : {img_half.shape}")
    print(f"Shape ảnh 1/4 : {img_quarter.shape}")
```

```python
if img is not None:
    out_half_path = 'output_images/bai3/output_half.jpg'
    out_quarter_path = 'output_images/bai3/output_quarter.jpg'
    
    cv2.imwrite(out_half_path, img_half)
    cv2.imwrite(out_quarter_path, img_quarter)
    
    size_original = os.path.getsize(img_path) / 1024.0
    size_half = os.path.getsize(out_half_path) / 1024.0
    size_quarter = os.path.getsize(out_quarter_path) / 1024.0

    print("--- SO SÁNH DUNG LƯỢNG FILE (KB) ---")
    print(f"Dung lượng ảnh gốc : {size_original:.2f} KB")
    print(f"Dung lượng ảnh 1/2 : {size_half:.2f} KB")
    print(f"Dung lượng ảnh 1/4 : {size_quarter:.2f} KB")
```

```python
if img is not None:

    img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    img_half_rgb = cv2.cvtColor(img_half, cv2.COLOR_BGR2RGB)
    img_quarter_rgb = cv2.cvtColor(img_quarter, cv2.COLOR_BGR2RGB)
    
    fig, axes = plt.subplots(1, 3, figsize=(15, 5))
    
    axes[0].imshow(img_rgb)
    axes[0].set_title(f'Ảnh Gốc\n{img.shape}')
    axes[0].axis('off')
    
    axes[1].imshow(img_half_rgb)
    axes[1].set_title(f'Ảnh 1/2\n{img_half.shape}')
    axes[1].axis('off')
    
    axes[2].imshow(img_quarter_rgb)
    axes[2].set_title(f'Ảnh 1/4\n{img_quarter.shape}')
    axes[2].axis('off')
    
    plt.tight_layout()
    plt.show()
```
