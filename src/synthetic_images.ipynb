# Synthetic Images Notebook


```python
import os
import cv2
import numpy as np
from IPython.display import display, Image
os.makedirs('output_images/bai4', exist_ok=True)
```

```python
height, width = 512, 512
gradient = np.zeros((height, width), dtype=np.uint8)
for i in range(width):
    gradient[:, i] = int(i / width * 255)
cv2.imwrite('output_images/bai4/gradient.png', gradient)
print('Gradient shape:', gradient.shape)
display(Image(filename='output_images/bai4/gradient.png'))
```

```python
canvas = np.zeros((height, width, 3), dtype=np.uint8)
stripe_h = height // 3
canvas[0:stripe_h, :] = (0, 0, 255)
canvas[stripe_h:2*stripe_h, :] = (255, 255, 255)
canvas[2*stripe_h:height, :] = (255, 0, 0)
cv2.imwrite('output_images/bai4/stripes.png', canvas)
print('Stripes shape:', canvas.shape)
display(Image(filename='output_images/bai4/stripes.png'))
```

```python
checker = np.zeros((height, width), dtype=np.uint8)
cell = 64
for r in range(0, height, cell):
    for c in range(0, width, cell):
        if ((r // cell) + (c // cell)) % 2 == 0:
            checker[r:r+cell, c:c+cell] = 255
cv2.imwrite('output_images/bai4/checkerboard.png', checker)
print('Checkerboard shape:', checker.shape)
display(Image(filename='output_images/bai4/checkerboard.png'))
```

```python
shapes = np.zeros((height, width, 3), dtype=np.uint8)
cv2.rectangle(shapes, (50, 50), (200, 200), (0, 255, 0), 3)
cv2.circle(shapes, (350, 150), 80, (0, 0, 255), -1)
cv2.ellipse(shapes, (256, 350), (120, 60), 0, 0, 360, (255, 255, 0), 2)
pts = np.array([[150, 400], [256, 300], [362, 400]], np.int32)
cv2.polylines(shapes, [pts], True, (255, 0, 255), 3)
cv2.line(shapes, (0, 0), (512, 512), (255, 255, 255), 1)
cv2.imwrite('output_images/bai4/shapes.png', shapes)
print('Shapes image shape:', shapes.shape)
display(Image(filename='output_images/bai4/shapes.png'))
```

```python
noise = np.random.randint(0, 256, (height, width, 3), dtype=np.uint8)
cv2.imwrite('output_images/bai4/noise.png', noise)
print('Noise shape:', noise.shape)
display(Image(filename='output_images/bai4/noise.png'))
```

```python
color_gradient = np.zeros((height, width, 3), dtype=np.uint8)
for i in range(width):
    color_gradient[:, i, 0] = int(i / width * 255)
for j in range(height):
    color_gradient[j, :, 1] = int(j / height * 255)
color_gradient[:, :, 2] = 128
cv2.imwrite('output_images/bai4/color_gradient.png', color_gradient)
print('Color gradient shape:', color_gradient.shape)
display(Image(filename='output_images/bai4/color_gradient.png'))
```

```python
circles = np.zeros((height, width, 3), dtype=np.uint8)
center = (width // 2, height // 2)
for r in range(20, 260, 30):
    color = tuple(np.random.randint(100, 256, 3).tolist())
    cv2.circle(circles, center, r, color, 2)
cv2.imwrite('output_images/bai4/concentric_circles.png', circles)
print('Concentric circles shape:', circles.shape)
display(Image(filename='output_images/bai4/concentric_circles.png'))
```

```python
grid = np.ones((height, width), dtype=np.uint8) * 255
for i in range(0, width, 32):
    cv2.line(grid, (i, 0), (i, height), 0, 1)
for j in range(0, height, 32):
    cv2.line(grid, (0, j), (width, j), 0, 1)
cv2.imwrite('output_images/bai4/grid.png', grid)
print('Grid shape:', grid.shape)
display(Image(filename='output_images/bai4/grid.png'))
```

```python
text_img = np.zeros((height, width, 3), dtype=np.uint8)
cv2.putText(text_img, 'OpenCV', (80, 280), cv2.FONT_HERSHEY_SIMPLEX, 2.5, (0, 255, 0), 4)
cv2.putText(text_img, 'Synthetic', (100, 380), cv2.FONT_HERSHEY_DUPLEX, 2, (0, 200, 255), 3)
cv2.imwrite('output_images/bai4/text.png', text_img)
print('Text image shape:', text_img.shape)
display(Image(filename='output_images/bai4/text.png'))
```
